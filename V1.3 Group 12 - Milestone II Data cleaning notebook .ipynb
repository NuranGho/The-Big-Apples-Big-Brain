### Extract and split data to multiple columns

```python
eladf['District Number'] = eladf['DBN'].str[0:2]
eladf['Borough Code'] = eladf['DBN'].str[2:3]
eladf['School Number'] = eladf['DBN'].str[3:]

mathdf['District Number'] = mathdf['DBN'].str[0:2]
mathdf['Borough Code'] = mathdf['DBN'].str[2:3]
mathdf['School Number'] = mathdf['DBN'].str[3:]
```

```python
mathdf.drop('DBN', axis=1, inplace = True)
eladf.drop('DBN', axis=1, inplace = True)
```

```python
mathdf.drop(['Unnamed: 0'], axis=1, inplace = True )
eladf.drop(['Unnamed: 0'], axis=1, inplace = True)
```

```python
mathdf.rename(columns = {'Number Tested':'Math Number Tested'}, inplace = True)
eladf.rename(columns = {'Number Tested':'ELA Number Tested'}, inplace = True)
```

```python
mathdf.rename(columns = {'Mean Scale Score':'Math Mean Scale Score'}, inplace = True)
eladf.rename(columns = {'Mean Scale Score':'ELA Mean Scale Score'}, inplace = True)
```

```python
eladf2 = eladf.loc[:, ["School Name", "Grade", "Year", "ELA Number Tested", "ELA Mean Scale Score"]]
```

```python
finaldf = mathdf.merge(eladf2, on=["School Name", "Grade", "Year"])
```

```python
finaldf.info()
```

```python
finaldf = finaldf.iloc[:, [0, 1, 2, 3,4,8,9,5,6,7]]
```

```python
finaldf
```

```python
year_dim=finaldf.copy()
year_dim=year_dim[['Year']]
```

```python
unique_row = ['Year']
year_dim = year_dim.drop_duplicates (subset = unique_row, keep = 'first')
year_dim = year_dim.reset_index(drop = True)
year_dim
```

```python
surrogate_key_start = 10
year_dim.insert(0, "year_id", range(surrogate_key_start, surrogate_key_start + len(year_dim)))
year_dim
```

```python
finaldf=finaldf.merge(year_dim,left_on=unique_row,right_on=unique_row,how='left')
```

```python
boroughs = {'M': 'Manhattan', 'X': 'Bronx', 'K': 'Brooklyn', 'Q': 'Queens', 'R': 'Staten Island'}

finaldf['Borough'] = finaldf['Borough Code'].map(boroughs)
```

```python
location_dim=finaldf.copy()
location_dim=location_dim[['District Number','Borough','School Number']]
```

```python
unique_row = ['District Number','Borough','School Number']
location_dim = location_dim.drop_duplicates (subset = unique_row, keep = 'first')
location_dim = location_dim.reset_index(drop = True)
location_dim
```

```python
surrogate_key_start = 10000
location_dim.insert(0, "Location_id", range(surrogate_key_start, surrogate_key_start + len(location_dim)))
location_dim
```

```python
finaldf=finaldf.merge(location_dim,left_on=unique_row,right_on=unique_row,how='left')
```

```python
grade_dim=finaldf.copy()
grade_dim=grade_dim[['Grade']]
```

```python
unique_row = ['Grade']
grade_dim = grade_dim.drop_duplicates (subset = unique_row, keep = 'first')
grade_dim = grade_dim.reset_index(drop = True)
grade_dim
```

```python
surrogate_key_start = 0
grade_dim.insert(0, "Grade_id", range(surrogate_key_start, surrogate_key_start + len(grade_dim)))
grade_dim
```

```python
finaldf=finaldf.merge(grade_dim,left_on=unique_row,right_on=unique_row,how='left')
```

```python
finaldf
```

```python
finaldf = finaldf.loc[~((finaldf['Math Mean Scale Score'] == 's') & (finaldf['ELA Mean Scale Score']=='s'))]
```

```python
finaldf = finaldf.replace(['s'], 0)
```

```python
grades_fact = finaldf[['year_id',
                       'Grade_id',
                       'Location_id',
                       'Math Number Tested',
                       'Math Mean Scale Score',
                       'ELA Number Tested',
                       'ELA Mean Scale Score']]
```
